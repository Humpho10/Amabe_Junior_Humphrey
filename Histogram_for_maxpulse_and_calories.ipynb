```python
#Data cleaning
import pandas as pd
df=pd.read_csv("data.csv")#df stands for data frame

#Empty values
#First check for empty values
print(df.isnull().sum())#This will give you the number of empty values in each column

```

```python
#Check wrong formats
print(df.dtypes)#This will give you the data type of each column
#check for non numeric values in numeric columns
for col in['Duration','Pulse','Maxpulse','Calories']:
    if not pd.to_numeric(df[col], errors='coerce').notnull().all():
        print(f"Column {col} has non-numeric values.")
        
        
#check for wrong data
for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']:
    if (df[col] < 0).any():
        print(f"Column {col} has negative values.")
        
#Duplicates
duplicates = df.duplicated()
if duplicates.any():
    print("Duplicates found:")
    print(df[duplicates])
else:
    print("No duplicates found.")        
        
```

```python
#1. Empty cells, don't delete, replace them with mean or median
for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']:
    df[col].fillna(df[col].mean(), inplace=True)
    
print(df.to_string())    
```

```python
#EDA-Exploratory Data Analysis
print(df.shape) #tells you the number of rows and columns
df.describe() #gives you the summary statistics of the data
```

```python
#Histogram for Maxpulse
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.histplot(df['Maxpulse'], kde=True, bins=30)
plt.title('Distribution of Maxpulse')


#Histogram for Calories
plt.subplot(2, 2, 2)
sns.histplot(df['Calories'], kde=True, bins=30)
plt.title('Distribution of Calories')

```
